```python
import graphviz
```

```python
layer_num = 7
node_list = [10, 7, 5, 3, 5, 7, 10]
nn = [['l%dn%d'%(layer_n, i) for i in range(node_list[layer_n])] for layer_n in range(layer_num)]
```

```python
g = graphviz.Digraph(name='00', directory='graph/00',
                     graph_attr={'rankdir': 'LR',
                                 'splines': 'ortho'})
for layer in nn:
    for node in layer:
        g.node(node, shape='circle', fontcolor='black', fontsize='5', style='filled', fillcolor='black')

for edge in [[(nodef, nodeb) for nodef in nn[i-1] for nodeb in nn[i]] for i in range(1,len(nn))]:
    g.edges(edge)

g.edge_attr.update(arrowhead='none')
g
```

```python
g = graphviz.Digraph(name='00', directory='graph/00',
                     graph_attr={'rankdir': 'LR',
                                 'splines': 'line'})
for layer in nn:
    for node in layer:
        g.node(node, shape='circle', fontcolor='black', fontsize='5', style='filled', fillcolor='black')

for edge in [[(nodef, nodeb) for nodef in nn[i-1] for nodeb in nn[i]] for i in range(1,len(nn))]:
    g.edges(edge)

g.edge_attr.update(arrowhead='none')
g
```

```python
def gen_fc_nn_nodes(layers_nodes=[20,15,10,5]):
    layer_num = len(layers_nodes)
    nn = [['l%dn%d'%(layer_n, i) for i in range(layers_nodes[layer_n])] for layer_n in range(layer_num)]
    return nn

def gen_fc_nn(nn):
    g = graphviz.Digraph(name='nn', directory='graph/nn',
                     graph_attr={'rankdir': 'LR',
                                 'splines': 'line',
                                 'overlap': 'Flase'})
    for layer in nn:
        for node in layer:
            g.node(node, shape='circle', fontcolor='black', fontsize='2', style='filled', fillcolor='black')

    for edge in [[(nodef, nodeb) for nodef in nn[i-1] for nodeb in nn[i]] for i in range(1,len(nn))]:
        g.edges(edge)

    g.edge_attr.update(arrowhead='none')
    return g
```

```python
gen_fc_nn(gen_fc_nn_nodes(layers_nodes=list(range(10,1,-1))))
```

```python
gen_fc_nn(gen_fc_nn_nodes(layers_nodes=list(range(8,1,-2)) + list(range(4,10,2))))
```
